```python
import re
```

```python
# splitting by whitespace

text = "I\m with you for always whenever you need my support, however, you'r staying in U.K so will connect over zoom" 
words = re.split(r'\W+', text)
```

```python
print(words)
```

```python
import string
```

```python
# split into words by whitespace
words = text.split()
words
```

```python
text = "I\m wi-t%h yo#u fo_r al$ways whe@@never y!ou n/eed m*y sup'port, howev'er, you'r stay%ing i$n U.K so will connect over zoom" 
words = text.split()
```

```python
words
```

```python
# prepare regex for char filtering

re_pun = re.compile('[%s]' % re.escape(string.punctuation))
re_pun
```

```python
# remove punctuation from each words

stripped = [re_pun.sub('',w) for w in words]
```

```python
print(stripped)
```

```python
# string.printable inverse of string.punctuation

re_print = re.compile('[^%s]' % re.escape(string.printable))
result = [re_print.sub('',w) for w in words]
```

```python
print(result)
```

# Normalizing case

```python
# split the words by white space
words = text.split()

# convert to lower case
words = [word.lower() for word in words]

print(words)
```
